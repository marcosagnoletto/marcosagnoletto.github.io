```python
u, v = var('u,v')
```

```python
assume(-1<v<1)
```

```python
assume(0<u<2*pi)
```

```python
mobius = ParametrizedSurface3D([(2+v*cos(u/2))*cos(u),(2+v*cos(u/2))*sin(u),v*sin(u/2)],[u,v],'mobius')
```

```python
A = mobius.plot((0.001,(2*pi)-0.001),(-0.999,0.999), color='lightblue')
```

```python
from sage.plot.plot3d.plot3d import axes
T = axes(1,1, color='black')
T = T + text3d('X',(1,0.2,0)) + text3d('Y',(0.2,1,0)) + text3d('Z',(0.2,0.2,1))
```

```python
(A + T).show(frame=false)
```

```python
a, b = var('a,b')
```

```python
assume(-1<a<1)
```

```python
assume(0<b<2*pi)
```

```python
mobius2 = ParametrizedSurface3D([(2+a*cos(b/2+pi/4))*cos(b+pi/2),(2+a*cos(b/2+pi/4))*sin(b+pi/2),a*sin(b/2+pi/4)],[b,a],'mobius2')
```

```python
B =  mobius2.plot((0.001,(2*pi)-0.001),(-0.999,0.999), color='lightgreen')
```

```python
(B + T).show(frame=false)
```

```python
(A + B + T).show(frame=false)
```
