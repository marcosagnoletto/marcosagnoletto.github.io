```python
u,v = var('u,v')
```

```python
assume(0<u<pi/2)
```

```python
assume(0<v<2*pi)
```

```python
sphere = ParametrizedSurface3D([sin(u)*cos(v),sin(u)*sin(v),cos(u)], [u, v],'sphere')
```

```python
A = sphere.plot((0,pi/2),(0,2*pi), color='lightblue')
```

```python
A.show()
```

```python
sphere2 = ParametrizedSurface3D([sin(u)*cos(v),sin(u)*sin(v),-cos(u)], [u, v],'sphere2')
```

```python
A2 = sphere2.plot((0,pi/2),(0,2*pi), color='lightblue')
```

```python
A2.show()
```

```python
(A+A2).show()
```

```python
g1 = [c[-1] for c in sphere.geodesics_numerical([sin(0.1)*cos(0.1),sin(0.1)*sin(0.1),cos(0.1)],[1,0],[0,2*pi,100])]
g2 = [c[-1] for c in sphere.geodesics_numerical([sin(0.1)*cos(0.1),sin(0.1)*sin(0.1),cos(0.1)],[0,1],[0,100,100])]
```

```python
(A + A2 + line3d(g1, color='red') + line3d(g2, color='blue')).show()
```
